# 🏀 Basketball Playbook Animation
Dieses Notebook visualisiert Spielzüge mit **animierten Spielerbewegungen**.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import json

# Lade Spielzug-Daten
with open('basketball_playbook.json', 'r') as f:
    spielzug = json.load(f)['spielzug']

# Spielfeldgröße (halb)
feld_x, feld_y = 14, 15  

# Matplotlib-Plot für Animation
fig, ax = plt.subplots()
ax.set_xlim(0, feld_x)
ax.set_ylim(0, feld_y)
ax.set_xticks([])
ax.set_yticks([])
ax.set_title('Basketball-Spielzug')

# Spieler als Punkte
spieler_punkte, = plt.plot([], [], 'ro', markersize=10)

def update(frame):
    x_coords = [pos[0] for pos in spielzug[frame]]
    y_coords = [pos[1] for pos in spielzug[frame]]
    spieler_punkte.set_data(x_coords, y_coords)
    return spieler_punkte,

ani = animation.FuncAnimation(fig, update, frames=len(spielzug), interval=1000, repeat=True)
plt.show()
```

### ⚙️ Anpassung:
- Die Spielzüge können in `basketball_playbook.json` geändert werden.
- Die Positionen der Spieler pro Sekunde anpassen.
- Mehr Bewegungen hinzufügen.
